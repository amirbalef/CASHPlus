```python
policy_algorithms = {}
policy_algorithms["PS_PFN_Mix"] = 1
policy_algorithms["PS_PFN_Pos"] = 1
policy_algorithms["PS_PFN_Neg"] = 1
policy_algorithms["PS_Max"] = 1
policy_algorithms["MaxUCB"] = 1
policy_algorithms["Rising_Bandit"] = 1

policy_algorithms["Random"] = 1

dataset_name = "synth_ablation"
```
